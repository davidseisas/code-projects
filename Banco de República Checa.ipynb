```python
from sklearn.impute import SimpleImputer
```

```python
imp_media = SimpleImputer(strategy="mean")
```

```python
columnas_imputar = ["population", "average_salary", "unemployment_rate", "num_committed_crimes"]
account_district[columnas_imputar] = imp_media.fit_transform(account_district[columnas_imputar])
```

```python
#remplazar los nulls en las columnas district_name y region con "UNKNOWN"
account_district = account_district.fillna("UNKNOWN")
```

```python
#Combinar trans y account_district en un nuevo DataFrame que se llame all_data
all_data = pd.merge(trans, account_district, on="account_id", how="left")
```

```python

all_data.info()
```

```python
#Crear una nueva columna account_open_year, que sera el año de la columna account_open_date
all_data["account_open_year"] = all_data["account_open_date"].dt.year
```

```python
all_data["account_open_year"].head()
```

```python
#Calcular la diferencia entre las columnas date(fecha de transacción) y account_open_date

all_data["date"] - all_data["account_open_date"]
```

```python
#Crear una nueva columna account_age_days que sera la diferencia en días entre las 
#columnas date(fecha de transacción) y account_open_date.

all_data["account_age_days"] = (all_data["date"] - all_data["account_open_date"])/pd.Timedelta(days=1)
```

```python
all_data
```

```python
#Crear una nueva columna "amount_category" dividiendo la columna "amount" en 3 caracteristicas
#del mismo tamaño y nombrar las caracteristicas como "cantidad_baja", "cantidad_media", "cantidad_alta".

all_data["amount_category"] = pd.qcut(all_data["amount"], q=3, labels=["low_amount", "medium_amount", "high_amount"])
```

```python
all_data["amount_category"].value_counts()
```

```python
#Crear la columna account_age_days_category que contenga la division de la columna account_age_days
#en 5 caracteristicas del mismo tamaño.

all_data["account_age_days_category"] = pd.cut(all_data["account_age_days"], bins=5)
```

```python
all_data["account_age_days_category"].value_counts()
```
